```python
from statsbombpy import sb
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
```

```python
world_cup_2022_matches = sb.matches(competition_id=43, season_id=106)
final_events = sb.events(match_id=3869685)
display(final_events)
```

```python
shots = final_events[["minute","shot_statsbomb_xg","team","location"]].dropna()
# display(shots)
```

```python
shots_ARG = shots.loc[shots["team"] == "Argentina"]
#print(shots_ARG)
shots_FRA = shots.loc[shots["team"] == "France"]
#print(shots_FRA)
```

```python
xg_ARG = shots_ARG["shot_statsbomb_xg"].sum()
xg_ARG = round(xg_ARG,2)
print(xg_ARG)

xg_FRA = shots_FRA["shot_statsbomb_xg"].sum()
xg_FRA = round(xg_FRA,2)
print(xg_FRA)
```

```python
local_shot_ARG = pd.DataFrame(shots_ARG["location"].to_list(), columns = ["x","y"])
local_shot_FRA = pd.DataFrame(shots_FRA["location"].to_list(), columns = ["x","y"])
#display(local_shot_ARG)
#display(local_shot_FRA)
```

```python
from todofcpy.visualization.graph import Field

field = Field()
plt = field.create_field(grass = "#FFFFFF", lines = "#C8102E")
plt.scatter(local_shot_ARG["x"], local_shot_ARG["y"], s = shots_ARG["shot_statsbomb_xg"]*150, c="#012169")
plt.title(xg_ARG)
plt.suptitle("Argentina xG")
plt.savefig("ArgentinaXG.png")
plt.show()

field2 = Field()
plt2 = field.create_field(grass = "#FFFFFF", lines = "#C8102E")
plt2.scatter(local_shot_FRA["x"], local_shot_FRA["y"], s = shots_FRA["shot_statsbomb_xg"]*150, c="#012169")
plt2.title(xg_FRA)
plt2.suptitle("France xG")
plt2.savefig("FranceXG.png")
plt2.show()
```
