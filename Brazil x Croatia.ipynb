```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mplsoccer.pitch import Pitch
from statsbombpy import sb
```

```python
#Pegando os dados de eventos da partida entre Brasil x Croácia
events = sb.events(match_id=3869420)
events.info()
events
```

```python
#Separando apenas os eventos de passe
pass_events = events[events['type'] == 'Pass']
pass_events
```

```python
#Removendo colunas indesejadas 1
colunas_a_manter = ['team', 'player','type', 'timestamp', 'location','pass_end_location','pass_height', 'pass_outcome']
pass_events = pass_events.drop(columns=[coluna for coluna in pass_events.columns if coluna not in colunas_a_manter])
```

```python
#Ordenando as colunas e preenchendo as linhas vazias do pass_outcome com passes completos
pass_events = pass_events[['timestamp', 'player','type','team', 'location', 'pass_end_location', 'pass_outcome','pass_height' ]]
pass_events['pass_outcome'].fillna('Complete', inplace=True)
pass_events
```

```python
#Separando os dados d localização para novas colunas
pass_events['x'] = pass_events['location'].apply(lambda loc: loc[0])
pass_events['y'] = pass_events['location'].apply(lambda loc: loc[1])
pass_events['end_x'] = pass_events['pass_end_location'].apply(lambda loc: loc[0])
pass_events['end_y'] = pass_events['pass_end_location'].apply(lambda loc: loc[1])
```

```python
#Reemovendo colunas indesejadas 2
colunas_a_manter = ['team', 'player','type', 'timestamp', 'x','y','end_x','end_y','pass_height', 'pass_outcome']
pass_events = pass_events.drop(columns=[coluna for coluna in pass_events.columns if coluna not in colunas_a_manter])
```

```python
pass_events
```

```python
#Separando em passes do Brasil e da Croácia
bra_pass_events = pass_events.loc[pass_events['team'] == 'Brazil']
cro_pass_events = pass_events.loc[pass_events['team'] == 'Croatia']
```

```python
 # Criando o campo
pitch = Pitch(pitch_color='grass', line_color='white')
fig, ax = pitch.draw(figsize=(16,9))


#separando os passes errados e certos
bra_complete_pass_events = bra_pass_events.loc[bra_pass_events['pass_outcome'] == 'Complete']
bra_incomplete_pass_events = bra_pass_events.loc[bra_pass_events['pass_outcome'] != 'Complete']

#Inserindo os dados no gráfico
pitch.arrows(
xstart=bra_complete_pass_events.x, ystart=bra_complete_pass_events.y,xend=bra_complete_pass_events.end_x,
yend=bra_complete_pass_events.end_y, width=1, ax=ax, headwidth=10, headlength=10, color='#cccc00',
label='Completed Passes')


pitch.arrows(
xstart=bra_incomplete_pass_events.x, ystart=bra_incomplete_pass_events.y, xend=bra_incomplete_pass_events.end_x,
yend=bra_incomplete_pass_events.end_y, width=0.8, ax=ax, headwidth=10, headlength=10, color='#990000',
label='Incomplete Passes')

#Titulo do gráfico
ax.set_title(f'Brazil Passes')

# Legenda
ax.legend(loc='upper left')
plt.show()
```

```python
# Criando o Campo
pitch = Pitch(pitch_color='grass', line_color='white')
fig, ax = pitch.draw(figsize=(16,9))

#separando os passes errados e certos
cro_complete_pass_events = cro_pass_events.loc[cro_pass_events['pass_outcome'] == 'Complete']
cro_incomplete_pass_events = cro_pass_events.loc[cro_pass_events['pass_outcome'] != 'Complete']

# Inserindo os passes ccompletos
pitch.arrows(
xstart=cro_complete_pass_events.x, ystart=cro_complete_pass_events.y, xend=cro_complete_pass_events.end_x,
yend=cro_complete_pass_events.end_y, width=1, ax=ax, headwidth=10, headlength=10, color='#cccc00',
label='Completed Passes')

#inserindo os passes incompletos
pitch.arrows(
xstart=cro_incomplete_pass_events.x, ystart=cro_incomplete_pass_events.y, xend=cro_incomplete_pass_events.end_x,
yend=cro_incomplete_pass_events.end_y, width=0.8, ax=ax, headwidth=10, headlength=10, color='#990000',
label='Incomplete Passes')

#Inserindo título
ax.set_title(f'Croatia Passes')

#inserindo legenda
ax.legend(loc='upper left')
plt.show()
```
