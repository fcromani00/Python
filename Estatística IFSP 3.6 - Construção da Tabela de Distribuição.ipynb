```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Dados.csv')
df = df[["Nome", "Altura (cm)"]].set_index("Nome")
df
```

```python
df = df.dropna(axis=0, how="any")
df.sort_values(by="Altura (cm)")
```

```python
df["Intervalo Altura"] = np.where(df['Altura (cm)'] <= 160, '150-160cm',
                         np.where(df['Altura (cm)'] <= 170, '161-170cm',
                                  np.where(df['Altura (cm)'] <= 180, '171-180cm',
                                           np.where(df['Altura (cm)'] <= 190, '181-190cm', '191-200cm'))))

# Criação da coluna "Frequência"
df['Frequência'] = 1
df
```

```python
tabela_frequencia_relativa = df.drop('Altura (cm)', axis=1, inplace=True)
tabela_frequencia_relativa = df.groupby('Intervalo Altura').sum('Frequência')
tabela_frequencia_relativa
tabela_frequencia_relativa.to_csv('tabela_frequencia_relativa.csv')
```

```python
tabela_grafico = tabela_frequencia_relativa.reset_index()
#tabela_grafico.rename(columns={'Frequência': 'Frequência Total'}, inplace=True)
tabela_grafico
```

```python
#gráfico
plt.figure(figsize=(8, 5))

barras = plt.bar(tabela_grafico['Intervalo Altura'], tabela_grafico['Frequência'], color='lightgreen')

barras[2].set_color('Green')

plt.bar_label(barras, fontsize=12, fontweight='bold', color='white', padding=-15)

plt.title('Altura de alunos ADS 2022', fontweight='bold', fontsize=15)

plt.box(False)
plt.yticks([])
plt.tick_params(axis='x', length=0)

plt.savefig('tabela_frequencia_relativa.png', bbox_inches='tight', dpi=300)

plt.show()
```
