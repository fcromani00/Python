## ETFs - Web Scraping + Análise de Dados

O projeto a seguir tem como objetivo de coletar dados sobre os ETFs presentes no site https://www.etf.com/etfanalytics/etf-finder e organizar uma base de dados. Também será feita uma análise das ETFs no final do projeto.



```python
# Importando bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
# Criando navegador e entrando no site https://www.etf.com/etfanalytics/etf-finder

#driver = webdriver.Chrome(service = Service(ChromeDriverManager())
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install())) #utilize esta linha caso esteja utilizando o cóodigo pela primeira vez

url = "https://www.etf.com/etfanalytics/etf-finder"

driver.get(url)
```

```python
# Apertando no botão de 100 linhas por página com dados de ETFs

time.sleep(5)

botao_100 = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[1]/div/div[4]/button/label/span''')

driver.execute_script('arguments[0].click();',botao_100)
```

```python
# Coletando o número de páginas com dados de ETFs

numero_paginas = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/label[2]''')

numero_paginas = numero_paginas.text.replace('of ','')

numero_paginas = int(numero_paginas)

print(numero_paginas)
```

```python
# Loop que coleta dados presentes na página de "Fund Basics", e passa para a próxima página

lista_de_tabela_por_pagina = []

for pagina in range(0,numero_paginas):
    
    tabela = driver.find_element('xpath', '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')
    
    html_tabela = tabela.get_attribute('outerHTML')
    
    tabela_final = pd.read_html(html_tabela)[0]
    
    lista_de_tabela_por_pagina.append(tabela_final)
                                                    
    botao_passar_pagina = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')
    
    driver.execute_script('arguments[0].click()',botao_passar_pagina)
    
base_de_dados_completa = pd.concat(lista_de_tabela_por_pagina)

display(base_de_dados_completa)
```

```python
# Após a coleta de dados na aba "Fund Basics", vamos coletar na aba "Performance"

# Clicando na aba performance

botao_aba = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/ul/li[2]''')

driver.execute_script('arguments[0].click();',botao_aba)
```

```python
# Clicando no botão de voltar página, para coletar desde o início

for pagina in range(0,numero_paginas):
    
    botao_voltar_pagina = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[1]''')
    
    driver.execute_script('arguments[0].click();',botao_voltar_pagina)
```

```python
# Loop que coleta os dados da página de "Performance", e passa para a próxima página

lista_de_tabela_por_pagina = []

for pagina in range(0,numero_paginas):
    
    tabela = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div''')
    
    html_tabela = tabela.get_attribute('outerHTML')
    
    tabela_final = pd.read_html(html_tabela)[0]
    
    lista_de_tabela_por_pagina.append(tabela_final)
    
    botao_passar_pagina = driver.find_element('xpath','''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')
    
    driver.execute_script('arguments[0].click();',botao_passar_pagina)
    
base_de_dados_performance = pd.concat(lista_de_tabela_por_pagina)
    
base_de_dados_performance
```

```python
# Fechando o navegador

driver.quit()
```

```python
# Removendo o index das tabelas, e colocando o "Ticker" delas como index

base_de_dados_completa = base_de_dados_completa.set_index('Ticker')

base_de_dados_performance = base_de_dados_performance.set_index('Ticker')

# Para a análise serão utilizados apenas os dados de 3, 5 e 10 anos, portanto será utilizada apenas a performance destes períodos

base_de_dados_performance = base_de_dados_performance[[ '3 Years', '5 Years', '10 Years']]
```

```python
# Tratando os dados de performance
# Transformando os ""--" em valores nulos e os removendo da análise (pois serão analisados ETFs com presença nas colunas de 1, 5 e 10 anos)

base_de_dados_performance = base_de_dados_performance.replace('--', pd.NA)

base_de_dados_performance = base_de_dados_performance.dropna()
```

```python
# Removendo o caracter "%" e transformando os dados em float

base_de_dados_performance['3 Years'] = base_de_dados_performance['3 Years'].str.rstrip('%').astype(float)/100

base_de_dados_performance['5 Years'] = base_de_dados_performance['5 Years'].str.rstrip('%').astype(float)/100

base_de_dados_performance['10 Years'] = base_de_dados_performance['10 Years'].str.rstrip('%').astype(float)/100

base_de_dados_performance
```

```python
# Juntando os dados das abas "Fund Basics" e "Performance"

base_de_dados_final = base_de_dados_completa.join(base_de_dados_performance, how = 'inner')

base_de_dados_final
```

```python
# Removendo os ETFs alavancados da amostra.
# Um ETF alavancado é um fundo negociado em bolsa que utiliza empréstimos para aumentar o potencial de retorno do investimento.

base_de_dados_final = base_de_dados_final[~base_de_dados_final['Segment'].str.contains("Leveraged")]

base_de_dados_final
```

```python
# Fazer um rank de rentabilidade por cada período

base_de_dados_final['rank_3_anos'] = base_de_dados_final['3 Years'].rank(ascending = False)
base_de_dados_final['rank_5_anos'] = base_de_dados_final['5 Years'].rank(ascending = False)
base_de_dados_final['rank_10_anos'] = base_de_dados_final['10 Years'].rank(ascending = False)

base_de_dados_final
```

```python
base_de_dados_final['Rank Final'] = base_de_dados_final['rank_3_anos'] + base_de_dados_final['rank_5_anos'] + base_de_dados_final['rank_10_anos']

melhores_etfs = base_de_dados_final.sort_values(by = 'Rank Final')

melhores_etfs.head(10)
```
