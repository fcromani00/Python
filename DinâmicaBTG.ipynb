```python
df = pd.read_csv('Pasta1.csv', sep=';')
df.info()
df
```

```python
#Nível 1: calcular o volume que cada cliente negociou de dólar
#OBS: só somar e agrupar, sem precisar separar por posição. Somar os dados da coluna AMOUNT (quantidade de dólar operado)

df.groupby(['Client'])['Amount'].sum()
```

```python
#Nível 2: calcular o quanto cada cliente negociou de dólar, separando por posição (SELL e BUY)

df.groupby(['Client','Position'])['Amount'].sum()
```

```python
#Nível 3: calcular o preço médio da moeda, conforme operação SELL e BUY
#Primeiro vai precisar formatar a coluna de reais
```

```python
df['Total Price BRL'] = df['Total Price BRL'].str.replace(',', '.')
df
```

```python
df['Total Price BRL'] = pd.to_numeric(df['Total Price BRL'])
df.info()
df
```

```python
#Nível 3: calcular o preço médio da moeda, conforme operação SELL e BUY
df['Amount'] = df['Amount']/ 100
df['Total Price BRL'] = df['Total Price BRL']/ 100

df['Valor Dólar'] = df['Total Price BRL']/df['Amount']
df
```

```python
media_dolar = df['Valor Dólar'].mean()
print(media_dolar)
```

```python
#Nível 4: calcular o preço médio da moeda, por cliente, separando por SELL e BUY
df.groupby(['Client','Position'])['Valor Dólar'].mean()
```
